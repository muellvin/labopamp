```python
#working with impedances
r = 560e3
R = 100e3
C = 10e-9
K = 3.3/np.sqrt(2)
alpha=-0.085

b = lambda om: 1/(1j*om*R*C + R/r)
x = lambda um: -K* b(um)**2/(1+b(um)**2-b(omega)*alpha/10)
x_withoutr = lambda um: 1/((um*R*C)**2 - 1)*K

x_values = []
x_withoutr_values = []

for omega in measurement[:,0]:
    x_values.append(abs(    x(2*np.pi*omega)   ))
    x_withoutr_values.append(0.01*abs(    x_withoutr(2*np.pi*omega)   ))
    
plt.scatter(measurement[:,0], measurement[:,1], label='measurement')
plt.plot(measurement[:,0], x_values, label='algebraic solution')
#plt.plot(measurement[:,0], x_withoutr_values, label='theo')

plt.vlines(152, min(measurement[:,1]),max(measurement[:,1]), color='red', label='freq of trans. osc. = 152 Hz')
#plt.vlines(159, min(measurement[:,1]),max(measurement[:,1]), color='red', label='freq of trans. osc.')

plt.xlabel('Frequency [kHz]')
plt.ylabel('RMS [V]')
plt.title('Resonance Curve')
plt.legend()
plt.savefig('resonance.png',dpi=150)
```

```python
from scipy.optimize import curve_fit

# Lorentzian function
def lorentzian(f, A, f0, gamma, C):
    return A *gamma**2/((f-f0)**2+gamma**2) + C


initial_guess = [8, 150, 3, 0.5]  
popt, pcov = curve_fit(lorentzian, measurement[:,0], measurement[:,1], p0=initial_guess)

A_fit, f0_fit, gamma_fit, C_fit = popt
print("Fitted parameters (Lorentzian):")
print(f"A = {A_fit:.3f}, f0 = {f0_fit:.3f}, gamma = {gamma_fit:.3f}, C = {C_fit:.3f}")

# Plot the data and the fitted curve
plt.figure(figsize=(8, 6))
plt.scatter(measurement[:,0], measurement[:,1], s=10, color='blue', label='Noisy data')
plt.plot(measurement[:,0], lorentzian(measurement[:,0], *popt), color='red', label='Lorentzian fit')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Lorentzian Fit to Resonance Curve')
plt.grid()
plt.show()
```
