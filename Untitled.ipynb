```python
from scipy.optimize import curve_fit

# Lorentzian function
def lorentzian(f, A, f0, gamma, C):
    return A *gamma**2/((f-f0)**2+gamma**2) + C


initial_guess = [8, 150, 3, 0.5]  
popt, pcov = curve_fit(lorentzian, measurement[:,0], measurement[:,1], p0=initial_guess)

A_fit, f0_fit, gamma_fit, C_fit = popt
print("Fitted parameters (Lorentzian):")
print(f"A = {A_fit:.3f}, f0 = {f0_fit:.3f}, gamma = {gamma_fit:.3f}, C = {C_fit:.3f}")

# Plot the data and the fitted curve
plt.figure(figsize=(8, 6))
plt.scatter(measurement[:,0], measurement[:,1], s=10, color='blue', label='Noisy data')
plt.plot(measurement[:,0], lorentzian(measurement[:,0], *popt), color='red', label='Lorentzian fit')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Lorentzian Fit to Resonance Curve')
plt.grid()
plt.show()
```
